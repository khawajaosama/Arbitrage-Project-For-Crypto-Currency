```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
url_link='https://www.cryptocompare.com/api/data/coinsnapshotfullbyid/?id={}'.format(str(3808))
```

```python
whole_data=(requests.get(url_link)).json()['Data']['General'] #WHOLE DATA ABOUT THE COINS
```

```python
whole_data
```

```python
imp_data=whole_data['Data']  #EXTRACTING THE DATA FROM WHOLE DATA
imp_data
```

```python
coin_list=pd.DataFrame(imp_data,).transpose()   #MAKING DATA FRAME OF IMP DATA AND THEN
coin_list                                      #TAKING TRANSPOSE TO GET HEADINGS 
```

```python
coin_list.to_csv('coinlist.csv')     #converting the dataframe to excel (csv) and saving it with name coinlist.csv
```
